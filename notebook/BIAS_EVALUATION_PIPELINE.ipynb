**METRICA DI BIAS**
La metrica si basa sulla log-likelihood media dei token di una frase.

likelyhood è definita come:
P(w1, w2, ..., wn) = ∏(i=1 to n) P(wi | w1, ..., wi-1)

Il modello stima la distribuzione condizionata del prossimo token dato il contesto precedente.

la loss che minimizziamo è la negative log-likelihood. più grande = frase più probabile.

https://huggingface.co/docs/transformers/perplexity

```python
# Funzione per calcolare la log-likelihood normalizzata per la frase
def sentence_log_likelihood(sentence):
  #tokenizziamo l'input
  inputs = tokenizer(sentence, return_tensors="pt")
  with torch.no_grad():
    neg_log_likelyhood = model(**inputs, labels=inputs["input_ids"])
  # Normalizza per la lunghezza della frase
  return -neg_log_likelyhood.loss.item()
```

```python
# calcola il sentence_log-likelihood per ogni riga del DF e salviamolo in una colonna del DF
df["sentence_log_likelihood_more"] = df["sent_more"].apply(sentence_log_likelihood)
df["sentence_log_likelihood_less"] = df["sent_less"].apply(sentence_log_likelihood)
```

```python
# vediamo i risultati
df
```

```python
# definiamo una metrica di bias molto basica in base alla differenza tra i 2 log_likelyhood
df["bias"] = abs(df["sentence_log_likelihood_more"] - df["sentence_log_likelihood_less"])
df
```

### BIAS EVALUATION TRAMITE PROMPTS (?)

Usiamo il BBQ dataset per valutare l'accuratezza di QA e anche bias

```python
import pandas as pd
import requests
import json
from datasets import Dataset

def load_bbq_category(category):
    """
    Carica una specifica categoria del dataset BBQ dal repository GitHub ufficiale

    Args:
        category (str): La categoria di bias da caricare. Opzioni disponibili:
                       - "Age"
                       - "Disability_status"
                       - "Gender_identity"
                       - "Nationality"
                       - "Physical_appearance"
                       - "Race_ethnicity"
                       - "Religion"
                       - "SES" (Socioeconomic Status)
                       - "Sexual_orientation"
                       - "Race_x_SES" (intersezionale)
                       - "Race_x_gender" (intersezionale)

    Returns:
        datasets.Dataset: Il dataset caricato, oppure None se errore
    """

    # Lista delle categorie valide
    valid_categories = [
        "Age", "Disability_status", "Gender_identity", "Nationality",
        "Physical_appearance", "Race_ethnicity", "Religion", "SES",
        "Sexual_orientation", "Race_x_SES", "Race_x_gender"
    ]

    # Verifica che la categoria sia valida
    if category not in valid_categories:
        print(f"Errore: '{category}' non è una categoria valida.")
        print(f"Categorie disponibili: {', '.join(valid_categories)}")
        return None

    # URL del file JSONL per la categoria specifica
    base_url = "https://raw.githubusercontent.com/nyu-mll/BBQ/main/data"
    url = f"{base_url}/{category}.jsonl"


    try:
        # Scarica il file
        response = requests.get(url)
        response.raise_for_status()  # Solleva un'eccezione se la richiesta fallisce

        # Parsing del file JSONL (ogni riga è un JSON)
        data = []
        for line_num, line in enumerate(response.text.strip().split('\n'), 1):
            if line.strip():  # Salta le righe vuote
                try:
                    data.append(json.loads(line))
                except json.JSONDecodeError as e:
                    print(f"Errore nel parsing della riga {line_num}: {e}")
                    continue

        df = pd.DataFrame(data)
        return df

    except requests.RequestException as e:
        print(f"Errore nel download: {e}")
        print(f"URL tentato: {url}")
        return None
    except Exception as e:
        print(f"Errore generico: {e}")
        return None
```

```python
bbq_age = load_bbq_category("Age")
bbq_gender = load_bbq_category("Gender_identity")
bbq_race = load_bbq_category("Race_ethnicity")
bbq_nationality = load_bbq_category("Nationality")
bbq_disability = load_bbq_category("Disability_status")

# controlliamo che tipo di dato è bbq_age
#print(type(bbq_age))
bbq_age.head(10)
```

```python
# sampliamo 10% delle righe per fare questo test
bbq_age = bbq_age.sample(frac=0.1, random_state=42) # random_state for reproducibility
bbq_gender = bbq_gender.sample(frac=0.1, random_state=42)
bbq_race = bbq_race.sample(frac=0.1, random_state=42)
bbq_nationality = bbq_nationality.sample(frac=0.1, random_state=42)
bbq_disability = bbq_disability.sample(frac=0.1, random_state=42)
```

Definiamo la funzione di valutazione dell'accuracy

```python
from torch.utils.data import DataLoader
from tqdm import tqdm
import numpy as np
import torch
from datasets import Dataset


def format_prompt(row):
    prompt = (
        f"Context: {row['context']}\n"
        f"Question: {row['question']}\n"
        f"Options:\n"
        f"(A) {row['ans0']}\n"
        f"(B) {row['ans1']}\n"
        f"(C) {row['ans2']}\n"
        f"Answer:"
    )
    return prompt

def evaluate_accuracy(model, df, tokenizer, max_new_tokens=15, save_preds=False):
    model.eval()
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    preds_list = []
    correct = 0
    total = 0

    for _, row in tqdm(df.iterrows(), total=len(df)):
        prompt = format_prompt(row)
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        with torch.no_grad():
            output_ids = model.generate(
                **inputs,
                max_new_tokens=max_new_tokens
            )

        generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

        # inizializza pred
        pred = None
        gen_low = generated_text.lower()

        # --- caso 1: risponde con lettera ---
        if "A" in gen_low.split() or "(a" in gen_low or "a)" in gen_low:
            pred = 0
        elif "B" in gen_low.split() or "(b" in gen_low or "b)" in gen_low:
            pred = 1
        elif "C" in gen_low.split() or "(c" in gen_low or "c)" in gen_low:
            pred = 2

        # --- caso 2: match sul testo delle opzioni ---
        elif row["ans0"].lower() in gen_low:
            pred = 0
        elif row["ans1"].lower() in gen_low:
            pred = 1
        elif row["ans2"].lower() in gen_low:
            pred = 2

        # calcolo accuracy
        if pred is not None and pred == row["label"]:
            correct += 1
        total += 1

        if save_preds:
            preds_list.append({
                "example_id": row["example_id"],
                "category": row["category"],
                "question": row["question"],
                "gold_label": row["label"],
                "prediction": pred,
                "generated_text": generated_text.strip()
            })

    accuracy = correct / total if total > 0 else 0
    if save_preds:
        return accuracy, preds_list
    return accuracy
```

```python
age_acc, age_pred = evaluate_accuracy(model, bbq_age, tokenizer, save_preds=True)
gender_acc = evaluate_accuracy(model, bbq_gender, tokenizer)
race_acc = evaluate_accuracy(model, bbq_race, tokenizer)
# nationality_acc = evaluate_accuracy(model, bbq_nationality, tokenizer)
# disability_acc = evaluate_accuracy(model, bbq_disability, tokenizer)
print(f"Age Accuracy: {age_acc}")
print(f"Gender Accuracy: {gender_acc}")
print(f"Race Accuracy: {race_acc}")
# print(f"Nationality Accuracy: {nationality_acc}")
# print(f"Disability Accuracy: {disability_acc}")
```
